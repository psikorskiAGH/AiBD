```python
data = pd.read_csv("../Original Data/drinks.csv")
d_first = data.copy()
d_second = data.copy()

d_first = d_first.drop('total_litres_of_pure_alcohol', axis=1)
d_second = d_second.drop(
    ['beer_servings', 'spirit_servings', 'wine_servings'], 
    axis=1
)

print(d_first)
print(d_second)

d_first.to_csv('../Analysis Data/drinks_servings_by_country.csv')
d_second.to_csv('../Analysis Data/drinks_pure_alcohol_by_country.csv')
```
