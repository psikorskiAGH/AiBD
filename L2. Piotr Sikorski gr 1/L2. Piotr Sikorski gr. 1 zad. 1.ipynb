## Zadanie 10

Jakie kategorie filmów zostały wypożyczone przez klientów?

**Odp.: Wszystkie 16**

```python
df = pd.read_sql(
    "select category.name from category, film_category, inventory, rental "
    "where category.category_id = film_category.category_id "  # połącz nazwy kategorii z filmami
    "and film_category.film_id = inventory.film_id "  # połącz koszyk klienta z kategorią filmu
    "and inventory.inventory_id = rental.inventory_id "  # połącz wypożyczenie z koszykiem klienta
    "group by category.name",  # grupuj wg kategorii filmu
    con=connection)
df
```

## Zadanie 11

Znajdź wszystkie kategorie filmów wypożyczonych w Ameryce.

Zakładam, że chodzi o wypożyczenia dla osób mieszkających w Ameryce.

**Odp.: Wszystkie 16**

```python
df = pd.read_sql(
    "select category.name from category, film_category, inventory, rental, "
    
        "(select * "  # zapytanie podrzędne zwracające przyporządkowanie konsumenta do jego państwa
        "from customer, address, city, country "  # jak w zadaniu 7
        "where customer.address_id = address.address_id "
        "and address.city_id = city.city_id "
        "and city.country_id = country.country_id "
        "and country.country = 'United States') as subquery "

    "where category.category_id = film_category.category_id " # jak w zadaniu 10
    "and film_category.film_id = inventory.film_id "
    "and inventory.inventory_id = rental.inventory_id "
    # połącz zapytanie podrzędne z wypożyczeniem za pomocą id konsumenta
    "and subquery.customer_id = rental.customer_id "  
    "group by category.name",
    con=connection)
df
```

## Zadanie 12

Znajdź wszystkie tytuły filmów, w których grał: Olympia Pfeiffer lub Julia Zellweger lub Ellen Presley

```python
df = pd.read_sql(
    "select film.title, actor.first_name, actor.last_name "
    "from film, film_actor, actor "  # połącz tabele ze sobą
    "where film.film_id = film_actor.film_id "
    "and film_actor.actor_id = actor.actor_id "
    "and ("  # wykonaj filtrowanie dla każdej pary (imię i nazwisko) osobno i połącz za pomocą OR
        "(actor.first_name = 'Olympia' and actor.last_name = 'Pfeiffer') "
        "or (actor.first_name = 'Julia' and actor.last_name = 'Zellweger') "
        "or (actor.first_name = 'Ellen' and actor.last_name = 'Presley') "
    ")"
    , con=connection)
df
```
